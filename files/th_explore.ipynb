```python
import matplotlib.pyplot as plt
import nibabel as nib
import numpy as np
```

```python
filepath = "subj01_pet.nii.gz"
file = nib.load(filepath)
data = file.get_fdata()
print(data.shape)
```

```python
slx = data[:,:,142]
dens = np.amax(slx)
slx = slx/dens
th = 0.571
for i in range(512):
    for j in range(512):
        if slx[i,j]>th:
            slx[i,j]=0
plt.figure(figsize=(5,5), dpi=300)
plt.imshow(np.rot90(slx), cmap='gray')
plt.title("th="+str(th)+" max="+str(dens))
```
